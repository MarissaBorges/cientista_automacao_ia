# Projeto Prático 1

```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris_data = pd.read_csv(url,header=None, names=column_names)

print('Exibindo as 5 primeiras linhas do DataFrame')
print(iris_data.head())

print('\nGerando um sumário do DataFrame')
print(iris_data.info())

print('\nMostrando estatísticas descritivas das colunas numéricas')
print(iris_data.describe())
```

## Usando o Matplolib

### Gráfico de Histograma

```python
import matplotlib.pyplot as plt

iris_data.hist(figsize=(10,8))
plt.suptitle('Histogramas de recursos do conjunto de dados Iris')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Gráfico de Dispersão

```python
plt.figure(figsize=(8, 6))
plt.scatter(iris_data['sepal_length'], iris_data['sepal_width'])
plt.title('Comprimento e Largura')
plt.ylabel('Largura (cm)')
plt.xlabel('Comprimento (cm)')
plt.grid(True)
plt.show
```

### Gráfico de Caixa

```python
plt.figure(figsize=(10, 6))
iris_data.boxplot()
plt.suptitle('Gráfico de Caixa dos dados da Iris')
plt.ylabel('cm')
plt.show()
```

### Gráfico de Caixa Agrupados

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(x='class', y='petal_length', data=iris_data)
plt.title('Largura Pental pela Classe')
plt.xlabel('Classe')
plt.ylabel('Largura Pental (cm)')
plt.show()
```

## Análise de correlação

```python
correlacao_matriz = iris_data.corr(numeric_only=True)
print('Correlação de Matriz')
print(correlacao_matriz)

plt.Figure(figsize=(8,6))
sns.heatmap(correlacao_matriz, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor da Correlação de Matriz')
```
