# Exercícios de Matplotlib

### Importação

```python
# importação
import matplotlib.pyplot as plt
import numpy as np
```

### 1 - Crie um gráfico de linhas da função y = x^2 para valores de x entre -10 e 10.

```python
x = np.linspace(-10, 10, 100)
y = x ** 2

plt.plot(x, y)
plt.show()
```

### 2 - Crie um gráfico de dispersão de 100 pontos aleatórios com cores diferentes com base em suas coordenadas x e y.

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, c=(x + y))
plt.show()
```

### 3 - Crie um gráfico de barras mostrando as vendas de cada mês do ano.

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [1200, 1500, 1700, 1300, 1800, 1600, 2000, 2100, 1900, 2200, 2300, 2500]

plt.bar(meses, vendas)
plt.show()
```

### 4 - Crie um histograma de um conjunto de dados com 500 valores, mostrando a distribuição dos dados.

```python
dados = np.random.rand(50)

plt.hist(dados)
plt.show()
```

### 5 - Crie uma figura com dois subgráficos: um mostrando um gráfico de linhas e o outro mostrando um gráfico de dispersão.

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [1200, 1500, 1700, 1300, 1800, 1600, 2000, 2100, 1900, 2200, 2300, 2500]

figura= plt.figure(figsize=(30,4))
figura.suptitle('Título da Figura')
facecolor='blue'

figura.add_subplot(121)
plt.plot(meses, vendas)
plt.title('Grafico de Linhas')

figura.add_subplot(122)
plt.scatter(meses, vendas)
plt.title('Grafico de Disperção')

plt.show(figura)
```
